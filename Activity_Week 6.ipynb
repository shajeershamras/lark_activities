```python
#importing the libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
#to read the provided dataset from location
df = pd.read_csv(r"C:\Users\shajeer\Desktop\Dataset_w6.csv")
```

```python
df.head()
```

# 1.Number of Countries in the dataset

```python
count=len(df.country)
print("Total number of Countries : ",count)
```

# 2.Visualizations

```python
plt.figure(figsize=(6,4))
sns.barplot(x='country',y='Agricultural production index ',data=df.sort_values('Agricultural production index ',ascending=False).head(10),color='red')
plt.title('Top 10 countries based on Agricultural production index',fontsize=15)
plt.xlabel('Country',fontsize=15);
plt.ylabel('Agricultural production index',fontsize=15);
plt.xticks(rotation=90);
plt.show();
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='country',y='GDP per capita (current US$)',data=df.sort_values('GDP per capita (current US$)',ascending=False).head(10),color='green')
plt.title('Top 10 countries based on GDP per capita (current US$)',fontweight='bold',fontsize=15);
plt.xlabel('Country',fontsize=15);
plt.ylabel('GDP per capita (current US$)',fontsize=15);
plt.xticks(rotation=90);
plt.show();
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Surface area (km2)'],df['Population in thousands (2017)'],s=10,c='red');
plt.title('PLOT OF surfacearea VS Population in thousands (2017)',fontsize=15);
plt.xlabel('surface area');
plt.ylabel('Population in thousands (2017)');
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df['Region'],df['International trade: Imports (million US$)'],s=10,c='red');
plt.title('PLOT OF Region VS International trade: Imports (million US$)',fontsize=10);
plt.xticks(rotation=90);
plt.xlabel('Region');
plt.ylabel('International trade: Imports (million US$)');
```

# 3.Countries with respect to region

```python
df.groupby('Region')[['country']].count().T
```

# 4.A.highest GDP per capita

```python
top=df.groupby('Region')[['GDP per capita (current US$)']].mean()
top.sort_values(by='GDP per capita (current US$)',ascending=False).T
```

# 4.B.Highest international trade

```python
top1=df.groupby('Region')[['International trade: Imports (million US$)']].mean()
top1.sort_values(by='International trade: Imports (million US$)',ascending=False).T
```

# 5.Region vs GDP per capita

```python
plt.figure(figsize=(8,4))
plt.scatter(df['Region'],df['GDP per capita (current US$)'],s=10,c='green');
plt.title('PLOT OF Region VS GDP per capita',fontsize=10);
plt.xticks(rotation=90);
plt.xlabel('Region');
plt.ylabel('GDP per capita');
```
