```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mdb=pd.read_csv(r'C:\Users\user\Desktop\Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv')
mdb.head()
```

# Churn count comparison with respect to gender

```python
plt.figure(figsize=(5,5))
sns.countplot( x='gender',hue='Churn',data= mdb) 
plt.title('Churn count respect to the gender',fontweight="bold")
plt.grid(linestyle="--")  
plt.show()
```

# Number of female senior citizens in dataset

```python
sdb1=mdb[(mdb['gender']=='Female') & (mdb['SeniorCitizen']==1)]
sdb1['gender'].count()
plt.figure(figsize=(6,6))
sns.countplot( x='gender',hue='SeniorCitizen',data=sdb1) 
plt.title("Count of female senior citizens ")
```

# Comparison of 'tenure' with 'Total Charges

```python
plt.figure(figsize=(8,8))
plt.scatter(mdb['tenure'] , mdb['TotalCharges'] , s=10 , c='green')
plt.title("Tenure VS Total charges",fontweight="bold")
plt.xlabel('Tenure')
plt.ylabel('Total Charges')
```
