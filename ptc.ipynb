```python
#missing value handling
#filling categorical values with mode method
for i in ['advice_to_mgmt','location','summary']:
     train_data[i]=train_data.fillna(train_data[i].mode())
#filling numericaldianl values with me method
for i in ['score_1','score_2','score_3','score_4','score_5']:
     train_data[i]=train_data.fillna(train_data[i].median())
```
